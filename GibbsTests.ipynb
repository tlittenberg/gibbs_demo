```python
import numpy as np
from chainconsumer import ChainConsumer
```

```python
standard_0 = np.loadtxt('standard_mcmc_0.dat')
standard_1 = np.loadtxt('standard_mcmc_1.dat')
standard_2 = np.loadtxt('standard_mcmc_2.dat')

parameters = ['x','y','z']

c = ChainConsumer()
c.add_chain(standard_0,parameters=parameters,name='standard trial 1')
c.add_chain(standard_1,parameters=parameters,name='standard trial 2')
c.add_chain(standard_2,parameters=parameters,name='standard trial 3')

c.configure(bar_shade=False,smooth=0,colors=['#1b9e77','#d95f02','#7570b3'],sigmas=[0,1,2,3],shade=False, shade_alpha=0.2)
fig = c.plotter.plot()
fig.set_size_inches(3 + fig.get_size_inches())
```

```python
standard_0 = np.loadtxt('standard_mcmc_0.dat')
parallel = np.loadtxt('parallel_gibbs.dat')
seriel = np.loadtxt('serial_gibbs.dat')

parameters = ['x','y','z']

c = ChainConsumer()
c.add_chain(standard_0,parameters=parameters,name='standard trial 1')
c.add_chain(parallel,parameters=parameters,name='parallel gibbs')
c.add_chain(seriel,parameters=parameters,name='serial gibbs')

c.configure(bar_shade=False,smooth=0,colors=['#1b9e77','#d95f02','#7570b3'],sigmas=[0,1,2,3],shade=False, shade_alpha=0.2)
fig = c.plotter.plot()
fig.set_size_inches(3 + fig.get_size_inches())
```
